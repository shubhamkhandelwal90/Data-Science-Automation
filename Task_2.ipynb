# 1. Python program to Create a excel file

```python
import xlsxwriter
import pandas as pd
import openpyxl
from openpyxl.chart import BarChart,Reference
import docx2txt
import re

number= int(input("Enter 1: Create Excel File\nEnter 2: Import Excel File\nEnter 3: Format data in excel sheet\nEnter 4: Format data in excel sheet\nEnter 5: Extract mobile no from MS word file and save into MS excel\nEnter choice of task:\n"))

#1. Create Excel File

if number== 1:
    workbook = xlsxwriter.Workbook('Engineering College Details.xlsx')
    worksheet = workbook.add_worksheet("Top Engg.Colleges")

    Colleges = (
            ['College Name','College City'],
            ['Indian Institute Of Technology (IIT)', 'Delhi'],
            ['Indian Institute Of Technology (IIT)', 'Bombay'],
            ['Indian Institute Of Technology (IIT)', 'Chennai'],
            ['Indian Institute Of Technology (IIT)', 'Kanpur'],
            ['Indian Institute Of Technology (IIT)', 'Kharagpur'],
             ['Indian Institute Of Technology (IIT)', 'Roorkee'],
        )
    row = 0
    col = 0

    for CollegeName, CollegeCity in (Colleges):
            worksheet.write(row, col, CollegeName)
            worksheet.write(row, col + 1, CollegeCity)
            row += 1
    workbook.close()


#2. Import Excel File
elif number == 2:
    EngineeringColleges = pd.read_excel('Engineering College Details.xlsx')
    EngineeringColleges.head()


#3.Format data in excel sheet
elif number == 3:
    workbook = xlsxwriter.Workbook('Engineering College Details_Formated.xlsx')
    worksheet = workbook.add_worksheet("Top Engg.Colleges")

    bold = workbook.add_format({'bold': True})

    worksheet.write('A1', 'College Name', bold)
    worksheet.write('B1', 'College City', bold)

    Colleges = (

            ['Indian Institute Of Technology (IIT)', 'Delhi'],
            ['Indian Institute Of Technology (IIT)', 'Bombay'],
            ['Indian Institute Of Technology (IIT)', 'Chennai'],
            ['Indian Institute Of Technology (IIT)', 'Kanpur'],
            ['Indian Institute Of Technology (IIT)', 'Kharagpur'],
             ['Indian Institute Of Technology (IIT)', 'Roorkee'],
        )

    row = 1
    col = 0

    for CollegeName, CollegeCity in (Colleges):
            worksheet.write(row, col, CollegeName)
            worksheet.write(row, col + 1, CollegeCity)
            row += 1

    workbook.close()


#4.Prepare excel charts
elif number == 3:
    wb = openpyxl.Workbook()
    sheet = wb.active

    for i in range(10):
        sheet.append([i])

    values = Reference(sheet, min_col = 1, min_row = 1,
    max_col = 1, max_row = 10)

    chart = BarChart()
    chart.add_data(values)

    chart.title = " BAR-CHART "
    chart.x_axis.title = " X_AXIS "
    chart.y_axis.title = " Y_AXIS "
    sheet.add_chart(chart, "E2")

    # save the file
    wb.save("barChart.xlsx")


#5. Extract mobile no from MS word file and save into MS excel
elif number == 5:
    doc_text = docx2txt.process('Number.docx')
    print(doc_text)

    number_word = re.compile('[98 97 96][0-9]{9}')
    number_match_word = number_word.finditer(doc_text)


    workbook_word = xlsxwriter.Workbook('Extract_no_from_docx.xlsx')
    worksheet_word = workbook_word.add_worksheet()

    row = 0
    col = 0

    for match in number_match_word:
        worksheet_word.write(row, col, match.group(0))

        row += 1

    workbook_word.close()

    Extracted_number = pd.read_excel('Extract_no_from_docx.xlsx',header=None,)
    Extracted_number.head()    
    
else:
    print("enter valid input")
```
